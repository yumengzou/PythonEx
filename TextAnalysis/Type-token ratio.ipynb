```python
def type_token(tokens):
    ans=len(set(tokens))/(float)(len(tokens))
    return ans
```

```python
def tokenize(rawstr):
    list=rawstr.split()
    punctuations="?.,;\/!'':-"
    ans=[s.lower().strip(punctuations) for s in list] ## list transformation
    return ans
```

```python
path="/Users/yumeng.zou/Documents/shakespeare/*.txt"
```

```python
import glob
filenames=glob.glob(path)
corpus=[tokenize(open(f,'r').read()) for f in filenames]
```

```python
import os
clean_filenames=[os.path.split(f)[1][:-4].replace("_"," ").capitalize() for f in filenames]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
xy=[(len(play), len(set(play))) for play in corpus]
x,y=zip(*xy)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x,y,s=100,alpha=.5)
for i,t in enumerate(clean_filenames):
    plt.text(x[i]+300,y[i]+20,t)
```
