```python
n=50
radius=10
for i in range(n):
    for j in range(n):
        d = ((i**2+j**2)**(1/2.0))
        bandNo= int(d)/radius
        if bandNo%2:
            print ".",
        else:
            print "+",
    print
```

```python
n=50
radius=10
for i in range(n):
    for j in range(n):
        d = (((i-(n/2.0))**2+(j-(n/2.0))**2)**(1/2.0))
        bandNo= int(d)/radius
        if bandNo%2:
            print ".",
        else:
            print "+",
    print
```
