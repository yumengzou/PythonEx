```python
import re
import glob,os
```

```python
def tokenize(rawstr):
    list=rawstr.split()
    punctuations="?.,;\/!'':-"
    ans=[s.lower().strip(punctuations) for s in list] ## list transformation
    return ans
```

```python
def vectorize(features, text):
    ans=[]
    for f in features:
        f_count=len([wd for wd in text if wd==f])
        ans.append(f_count)
    return ans
```

```python
path="/Users/yumeng.zou/Documents/shakespeare/*.txt"
filenames=glob.glob(path)
corpus=[tokenize(open(f,'r').read()) for f in filenames]
```

```python
def distance(c1,c2):
    ans=0
    for i, v in enumerate(c1):
        ans+=(v-c2[i])**2
    ans=ans**.5
    return ans  
## a sigma sum sign calls for a loop
```

```python
def re_vectorize(features, text):
    ans=[]
    for flst in features:
        regex="|".join(flst)
        count=len([wd for wd in text if re.match(regex,wd)])
        ans.append(count/(float)(len(text)))
    return ans
```

```python
cluster=["tragedy" if i==0 else "comedy" for i in kmeans.labels_]
```
